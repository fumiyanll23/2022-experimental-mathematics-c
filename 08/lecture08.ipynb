```python
# 必要なモジュールをインポートする
import sympy as sp
from sympy.plotting import plot
```

## 以下の常微分方程式の境界値問題の解 $y = y(x)$ を求める．

ここで $y' = y'(x) = \frac{dy(x)}{dx}, y'' = y''(x) = \frac{d^{2}y(x)}{dx^{2}}$ である．

### 1. $y'' = \sin{x}\ (0 < x < \pi), y(0) = y(\pi) = 0$.

解は $y = -\sin{x}$ ．

```python
# 常微分方程式{y'' = sinx (0<x<π), y(0) = y(π) = 0}を解く
# 変数xおよび関数yを定義する
x = sp.Symbol("x")
y = sp.Function("y")
# 常微分方程式を与える
eq = sp.Eq(y(x).diff(x, 2), sp.sin(x))
# 一般解を求める
ans_gen = sp.dsolve(eq)
print(f"一般解: {ans_gen.lhs} = {ans_gen.rhs}")
# 境界値条件代入して特殊解を求める
ans_sp = sp.dsolve(eq, ics={y(0): 0, y(sp.pi): 0})
print(f"特殊解: {ans_sp.lhs} = {ans_sp.rhs}")
# 特殊解のグラフを描画する
plot(ans_sp.rhs, (x, 0, sp.pi))
```

### 2. $y'' = -9y\ (0 < x < \pi)$.

### (i) 境界条件 $y(0) = y(\pi) = 0$.

解は $y = C_{1} \sin{3x}$ ($C_{1}$ は任意定数)．

```python
# 常微分方程式{y'' = -9y (0<x<π), y(0) = y(π) = 0}を解く
# 変数xおよび関数yを定義する
x = sp.Symbol("x")
y = sp.Function("y")
# 常微分方程式を与える
eq = sp.Eq(y(x).diff(x, 2), -9 * y(x))
# 一般解を求める
ans_gen = sp.dsolve(eq)
print(f"一般解: {ans_gen.lhs} = {ans_gen.rhs}")
# 境界値条件代入して特殊解を求める
ans_sp = sp.dsolve(eq, ics={y(0): 0, y(sp.pi): 0})
print(f"特殊解: {ans_sp.lhs} = {ans_sp.rhs}")
# 任意定数を指定して、特殊解のグラフを描画する
plotted = plot(
    sp.sin(3 * x), (x, 0, sp.pi), show=False, label="sin(3x) (C1=1)", legend=True
)
for c in range(2, 6):
    plotted.append(
        plot(
            c * sp.sin(3 * x),
            (x, 0, sp.pi),
            show=False,
            label=f"{c}*sin(3x) (C1={c})",
            legend=True,
        )[0]
    )
plotted.show()
```

### (ii) 境界条件 $y'(0) = y'(\pi) = 0$.

解は $y = C_{2} \cos{3x}$ ($C_{2}$ は任意定数)．

```python
# 常微分方程式{y'' = -9y (0<x<π), y'(0) = y'(π) = 0}を解く
# 変数xおよび関数yを定義する
x = sp.Symbol("x")
y = sp.Function("y")
# 常微分方程式を与える
eq = sp.Eq(y(x).diff(x, 2), -9 * y(x))
# 一般解を求める
ans_gen = sp.dsolve(eq)
print(f"一般解: {ans_gen.lhs} = {ans_gen.rhs}")
# 境界値条件代入して特殊解を求める
ans_sp = sp.dsolve(
    eq, ics={y(x).diff(x, 1).subs(x, 0): 0, y(x).diff(x, 1).subs(x, sp.pi): 0}
)
print(f"特殊解: {ans_sp.lhs} = {ans_sp.rhs}")
# 任意定数を指定して、特殊解のグラフを描画する
plotted = plot(
    sp.cos(3 * x), (x, 0, sp.pi), show=False, label="cos(3x) (C2=1)", legend=True
)
for c in range(2, 6):
    plotted.append(
        plot(
            c * sp.cos(3 * x),
            (x, 0, sp.pi),
            show=False,
            label=f"{c}*cos(3x) (C2={c})",
            legend=True,
        )[0]
    )
plotted.show()
```

### 3. $y'' + y - x = 0\ (0 < x < \pi), y(0) = y(\frac{\pi}{2}) = 0$.

解は $y = x - \frac{\pi}{2} \sin{x}$ ．

```python
# 常微分方程式{y''+y-x = 0 (0<x<π), y(0) = y(π/2) = 0}を解く
# 変数xおよび関数yを定義する
x = sp.Symbol("x")
y = sp.Function("y")
# 常微分方程式を与える
eq = sp.Eq(y(x).diff(x, 2) + y(x) - x, 0)
# 一般解を求める
ans_gen = sp.dsolve(eq)
print(f"一般解: {ans_gen.lhs} = {ans_gen.rhs}")
# 境界値条件代入して特殊解を求める
ans_sp = sp.dsolve(eq, ics={y(0): 0, y(sp.pi / 2): 0})
print(f"特殊解: {ans_sp.lhs} = {ans_sp.rhs}")
# 特殊解のグラフを描画する
plot(ans_sp.rhs, (x, 0, sp.pi))
```
